```python
import pandas as pd

df = pd.read_csv("../data/business_metrics.csv")
df
```

```python
df["revenue_mom_pct"] = df["revenue"].pct_change() * 100
df["avg_order_value"] = df["revenue"] / df["orders"]

df[["month","revenue","revenue_mom_pct","avg_order_value"]]
```

```python
latest = df.iloc[-1]
prev = df.iloc[-2]

revenue_change = (latest["revenue"] - prev["revenue"]) / prev["revenue"] * 100
orders_change = (latest["orders"] - prev["orders"]) / prev["orders"] * 100

print(f"Revenue MoM Change: {revenue_change:.2f}%")
print(f"Orders MoM Change: {orders_change:.2f}%")
```
